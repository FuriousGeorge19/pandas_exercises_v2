### Step 3. Assign it to a variable called chipo.

### Step 4. How many products cost more than $10.00?

```python
# convert item_price from string to float

chipo = chipo.assign(
    item_price = lambda x: x.item_price.str.replace('$', '').astype(float)
)
```

```python
chipo
```

```python
(
    chipo
    .drop_duplicates(subset=['item_name', 'quantity', 'choice_description'])
    .query('quantity == 1')
    .query('item_price > 10')
    .loc[:, 'item_price']
)
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
chipo_items =\
(
    chipo.query('quantity == 1')
    .drop_duplicates(subset=['item_name', 'choice_description'])
    .query('item_price > 10')
    .drop(['order_id', 'quantity', 'choice_description'], axis=1)
    .drop_duplicates()
)
    
```

### Step 6. Sort by the name of the item

```python
chipo_items.sort_values('item_name')
```
