### BONUS: Create your own question and answer it using a graph.

Look at the distribution of tip percentage vs sex, holding party size constant

```python
from  matplotlib.ticker import PercentFormatter

tips = tips.assign(
    tip_pct = lambda x: x.tip/x.total_bill
)

fig, ax = plt.subplots(figsize=(8,8))
sns.boxplot(data=tips, x='size', y='tip_pct', ax=ax, hue='sex')

ax.yaxis.set_major_formatter(PercentFormatter(1))
ax.set(
    title="Percentage Tip vs. Gender of Server, Holding Party Size Constant",
    xlabel="Party Size",
    ylabel="Percentage Tip"
)

plt.grid(axis='y')
```

I'm not sure splitting by party size shows a differentiation. Let's look at Tip Percentage vs. the Gender of the Server.

```python
fig, ax = plt.subplots(figsize=(8,8))
sns.boxplot(data=tips, x='sex', y='tip_pct', ax=ax, hue='sex')

ax.yaxis.set_major_formatter(PercentFormatter(1))
ax.set(
    title="Percentage Tip vs. Gender of Server",
    xlabel="Gender of the Server",
    ylabel="Percentage Tip"
)

plt.grid(axis='y')
```
